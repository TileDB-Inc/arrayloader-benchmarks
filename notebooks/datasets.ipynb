## [Papermill](https://papermill.readthedocs.io/en/latest/) params:

```python
census_uri = None
census_version = "2023-12-15"
collection_id = '283d65eb-dd53-496d-adb7-7570c7caa443'
```

```python
census = cellxgene_census.open_soma(uri=census_uri, census_version=census_version)
datasets = get_dataset_ids(census, collection_id)
len(datasets), datasets[:10]
```

List datasets from collection:

```python
ddf = get_datasets_df(census, collection_id)
ddf
```

Sort datasets in order of increasing cell count, compute cumulative sums:

```python
sagg = ddf.set_index('dataset_id').dataset_total_cell_count.sort_values().reset_index()
sagg['sum'] = sagg.dataset_total_cell_count.cumsum()
sagg
```

## 1e6 cutoffs
How many / Which datasets should be combined, to yield a total number of cells greater than each multiple of 1e6?

```python
mms = sagg['sum'] // 1e6
sagg[mms != mms.shift(1)]
```
